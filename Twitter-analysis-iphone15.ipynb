```python
# Create a function to get the subjectivity
def getSubjectivity (text):
    return TextBlob(text).sentiment.subjectivity

# Create a function to get the polarity
def getPolarity (text):
    return TextBlob(text). sentiment.polarity
```

```python
#Create two new columns
df ['Data_Subjectivity'] = df['tweet'].apply(getSubjectivity)
df['Data_Polarity'] = df['tweet'].apply(getPolarity)

#Show the new dataframe with the new columns
df
```

```python
#Evaluating positive, negative and neutral sentimaental analysis and adding it into the dataframe
def getAnalysis (x):
    if x < 0:
        return 'Negative'
    elif x == 0:
        return 'Neutral'
    else:
        return 'Positive'
```

```python
# Create a new Column with all conditions which we defined based on that give the output like negative, neutral and positive. 
df['Data_Analysis'] = df['Data_Polarity'].apply(getAnalysis)

#Print the Dataframe
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
Positive_sortedDF = df.sort_values(by=['Data_Polarity'])

for i in range(0, Positive_sortedDF.shape[0]):
    if(Positive_sortedDF['Data_Analysis'][i] == 'Positive'):
        print (str(i) + ' ) ' + Positive_sortedDF ['tweet'][i])
        print()
```

```python
# Gathering all the tweets and print it
Negative_sortedDF = df.sort_values (by=['Data_Polarity'])
for i in range(0, Negative_sortedDF.shape [0]):
    if(Negative_sortedDF [ 'Data_Analysis'][i] == 'Negative'):
        print (str(i) + ' ) ' + Negative_sortedDF ['tweet'][i])
        print()
```

```python
NeutralsortedDF = df.sort_values (by=['Data_Polarity'])
for i in range(0, NeutralsortedDF.shape [0]):
    if(NeutralsortedDF [ 'Data_Analysis'][i] == 'Neutral'):
        print (str(i) + ' ) ' + NeutralsortedDF ['tweet'][i])
        print()
```

### Visualizing the data

```python
# polarity and subjectivity plotting 
# Size of the plot
plt.figure(figsize=(8,6))

for i in range(0, df.shape[0]):
    plt.scatter (df['Data_Polarity'][i], df [ 'Data_Subjectivity'][i], color='Red' )

plt.title('Sentiment Analysis')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.show()
```

```python
# ploting pie chart to shows contribution of positive, Negative and Neutral tweets

# Each value was gotten by counting Negative_sortedDF, Negative_sortedDF and NeutralsortedDF
positive_value = 56
negative_value = 15
neutral_value = 34

labels = ['Positive', 'Negative', 'Neutral']

sizes = [positive_value, negative_value, neutral_value]

# Colors for each section
colors = ['green', 'gray', 'pink']

# Exploding the 1st slice (i.e., Positive)
explode = (0.1, 0, 0)

# Plotting the Pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Title of the Pie chart
plt.title('Sensitivity Analysis')

# Display the Pie chart
plt.show()
```
